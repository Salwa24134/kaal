

# SISTEM PERSAMAAN LINIER

Sistem Persamaan Linier (SPL) adalah kumpulan dari dua atau lebih persamaan yang memiliki variabel yang sama dan harus diselesaikan secara bersamaan. Dalam SPL, setiap persamaan berbentuk hubungan linier, artinya variabel-variabel dalam persamaan hanya memiliki pangkat satu dan tidak terdapat perkalian antar variabel.
Konsep Dasar SPL Dalam sistem ini, tujuan utamanya adalah mencari nilai dari variabel yang memenuhi semua persamaan yang ada. Misalnya, jika terdapat dua persamaan dengan dua variabel, kita harus menemukan nilai variabel tersebut sehingga kedua persamaan menjadi benar secara bersamaan.

Jenis-Jenis Solusi dalam SPL Memiliki satu solusi unik – Artinya hanya ada satu nilai variabel yang memenuhi semua persamaan dalam sistem. Memiliki banyak solusi (tak hingga) – Artinya ada banyak kemungkinan nilai variabel yang bisa memenuhi semua persamaan. Ini terjadi jika kedua persamaan menggambarkan garis yang sama dalam grafik. Tidak memiliki solusi – Artinya tidak ada nilai variabel yang bisa memenuhi semua persamaan dalam sistem. Ini terjadi jika kedua persamaan mewakili garis yang sejajar dan tidak pernah bertemu. Metode Penyelesaian SPL Ada beberapa metode umum yang digunakan untuk menyelesaikan SPL, di antaranya:

Metode Substitusi – Menggunakan salah satu persamaan untuk mengekspresikan satu variabel dalam bentuk variabel lainnya, lalu menggantinya ke persamaan lain. Metode Eliminasi – Mengeliminasi salah satu variabel dengan menjumlahkan atau mengurangkan persamaan sehingga tersisa satu variabel yang bisa diselesaikan. Metode Matriks – Menggunakan teknik matriks seperti eliminasi Gauss, metode invers matriks, atau aturan Cramer untuk menyelesaikan sistem secara lebih sistematis.

# Metode Eliminasi Gauss

  Metode eliminasi Gauss digunakan untuk menyelesaikan persamaan linear dengan cara mengubahnya ke bentuk eselon baris.
Pertama, sistem persamaan dituliskan dalam bentuk matriks yang disebut matriks Augmentasi.
Kemudian, dilakukan Operasi Baris Elementer(OBE) seperti menukar posisi baris, mengalikan baris  dengan suatu konstanta tidak nol, dan menjumlahkan atau mengurangkan kelipatan suatu baris ke baris lain.
Setelah itu, nilai dapat ditentukan dengan metode Substitusi Balik, yaitu menyelesaikan  persamaan dari baris terakhir ke baris pertama secara bertahap. Dengan demikian, solusi sistem persamaan dapat ditemukan secara terstruktur dan sistematis.

##### Diketahui sistem persamaan linear:
$$\begin{aligned}
x + y + z  &= 6 \\
2x - y + z &= 3 \\
3x + y - z &= 4
\end{aligned}$$

##### Penyelesaian :
1. Bentuk Matriks Augmentasi 
$$\begin{bmatrix}
   1 &  1 & 1 | 6  \\
   2 &-1 & 1| 3 \\
   3 &1& -1 | 4 
\end{bmatrix}$$

2. Eliminasi Gauss :

langkah 1. Nol-kan elemen di bawah elemen pivot pertama (1,1)
Gunakan baris pertama untuk mengeliminasi elemen di bawahnya menjadi nol:

##### hasilnya : 
$$\begin{bmatrix}
1 &  1 & 1  |  6  \\
0 & -3 & -1 | -9  \\
0 & -2 & -4 | -14
\end{bmatrix}$$

Langkah 2.Nol-kan elemen di bawah elemen pivot kedua adalah 1 di posisi (2,2).
Gunakan baris kedua untuk mengeliminasi baris ketiga:

##### hasilnya : 
$$\begin{bmatrix}
 1 &  1 & 1  |  6  \\
 0 & -3 & -1 | -9  \\
 0 &  0 & -2 | -6 
\end{bmatrix}$$

3. Substitusi Balik

Langkah 1 dari baris terakhir:
-2z = -4
  z =2
  
Langkah 2 subsitusi z ke baris kedua:
-3y - z = -9
-3y - 2 = -9
-3y = -7
y = 3

Langkah 3 substitusi y dan z ke baris pertama: 
x +y + z = 6
x + 3 + 2 = 6
x = 6 - 5
x = 1

jadi, solusi akhir dari sistem persamaan tersebut adalah 
x = 1, y = 3, z = 2

# BAB 5 Transformasi Matriks

### 5.1 Transformasi Matriks

Transformasi matriks adalah bentuk khusus dari transformasi linear, yang mengubah satu vektor menjadi vektor lain melalui perkalian matriks. Visualisasi ini sangat berguna terutama dalam bidang seperti grafika komputer. Definisi aljabar memungkinkan kita menggunakan matriks untuk mendefinisikan fungsi yang mengubah satu vektor menjadi vektor lain, seperti fungsi dalam matematika yang mengubah satu angka menjadi angka lain. Kemudian, kita bisa memvisualisasikan matriks dan perkalian matriks berdasarkan pengaruhnya terhadap vektor.

Diberikan sebuah matriks $A$ berukuran $m \times n$, kita bisa mendefinisikan fungsi yang menerima vektor kolom $x$ berukuran $n \times 1$ sebagai input, dan menghasilkan vektor kolom $Ax$ berukuran $m \times 1$ sebagai output, sesuai hubungan:

$$
\vec{y} = T(\vec{x}) = A\vec{x}
$$

Fungsi semacam ini dikenal sebagai transformasi matriks, dan merupakan contoh dari kelas fungsi yang lebih umum antara ruang vektor, yang disebut transformasi linear.

---

#### 5.1.1 Perkalian Matriks dengan Vektor

        Perkalian matriks dengan vektor dapat divisualisasikan secara geometris terutama dalam ruang dua dimensi (R²). Dalam contoh yang diberikan, beberapa vektor dikalikan dengan sebuah matriks 2×2, menghasilkan vektor baru dengan arah dan panjang yang berbeda. Ada vektor yang tetap searah setelah transformasi, sementara lainnya berubah arah. Hal ini menunjukkan bahwa matriks bertindak sebagai alat transformasi yang dapat memutar, memperbesar, memperkecil, atau bahkan menghilangkan vektor. Tidak semua matriks memengaruhi vektor dengan cara yang sama—beberapa memperpanjang semua vektor, beberapa memperpendek, dan ada juga yang hanya mengubah arah vektor tertentu. Untuk memahami bagaimana matriks memengaruhi vektor, digunakan pendekatan dengan menggabungkan operasi perkalian matriks dan penjumlahan vektor, yang juga divisualisasikan melalui grafik.
---

#### 5.1.2 Transformasi pada Bidang Kartesius

        Transformasi matriks tidak hanya memengaruhi satu vektor, tetapi juga dapat diterapkan ke seluruh bidang Kartesius. Karena sifat distributif dari perkalian matriks terhadap penjumlahan vektor, bidang ini akan mengalami perubahan secara konsisten—garis lurus tetap menjadi garis lurus, dan kurva tetap menjadi kurva meskipun bentuknya bisa berubah. Salah satu cara efektif untuk memahami dampak transformasi ini adalah dengan mengamati bagaimana matriks mengubah bentuk dari unit square (persegi satuan). Ketika keempat titik sudut dari persegi tersebut dikalikan dengan matriks, hasilnya menunjukkan bagaimana seluruh bidang akan berubah. Proses ini memberikan pemahaman visual yang kuat mengenai bagaimana transformasi matriks memengaruhi posisi dan bentuk dari kumpulan vektor.

---

#### 5.1.3 Kombinasi Penjumlahan dan Perkalian Matriks
        Interaksi antara penjumlahan vektor dan perkalian matriks menunjukkan bahwa sifat distributif tetap berlaku. Dalam contoh yang diberikan, penjumlahan dua vektor dilakukan terlebih dahulu, kemudian hasilnya dikalikan dengan matriks. Ternyata hasilnya sama dengan jika masing-masing vektor dikalikan terlebih dahulu, lalu dijumlahkan. Secara visual, hal ini diperjelas dengan grafik vektor dan bentuk paralelogram yang terbentuk. Kesimpulannya, transformasi matriks mempertahankan struktur aljabar dari vektor, khususnya sifat distributif terhadap penjumlahan. Pemahaman ini menjadi dasar untuk mengenali bahwa transformasi matriks adalah bentuk khusus dari transformasi linear.

# Pembuktian Transformasi Linier

##### Diberikan sebuah transformasi:

$$
T(v_1, v_2) = (v_1 + v_2, v_1)
$$

Tujuan kita adalah membuktikan bahwa $T$ merupakan **transformasi linier**.

---

##### Syarat Transformasi Linier

Agar suatu transformasi $T$ dikatakan **linier**, maka untuk sembarang vektor $\mathbf{a}, \mathbf{b} \in \mathbb{R}^2$ dan skalar $c \in \mathbb{R}$, harus berlaku:

1. **Aditivitas (Penjumlahan):**

   $$
   T(\mathbf{a} + \mathbf{b}) = T(\mathbf{a}) + T(\mathbf{b})
   $$

2. **Homogenitas (Perkalian skalar):**

   $$
   T(c \cdot \mathbf{a}) = c \cdot T(\mathbf{a})
   $$

---

#### Langkah Pembuktian Transformasi Linier

##### 1. Aditivitas (Penjumlahan)

Misalkan:

$$
\mathbf{a} = (a_1, a_2), \quad \mathbf{b} = (b_1, b_2)
$$

Hitung:

$$
T(\mathbf{a} + \mathbf{b}) = T(a_1 + b_1, a_2 + b_2) = ((a_1 + b_1) + (a_2 + b_2), a_1 + b_1)
$$

$$
= (a_1 + a_2 + b_1 + b_2, a_1 + b_1)
$$

Sementara:

$$
T(\mathbf{a}) + T(\mathbf{b}) = (a_1 + a_2, a_1) + (b_1 + b_2, b_1) = (a_1 + a_2 + b_1 + b_2, a_1 + b_1)
$$

Jadi:

$$
T(\mathbf{a} + \mathbf{b}) = T(\mathbf{a}) + T(\mathbf{b})
$$

**Syarat aditif terpenuhi**

---

##### 3 2. Homogenitas(Perkalian Skalar)

Misalkan $c \in \mathbb{R}$, dan $\mathbf{a} = (a_1, a_2)$

Hitung:

$$
T(c \cdot \mathbf{a}) = T(ca_1, ca_2) = (ca_1 + ca_2, ca_1) = c(a_1 + a_2, a_1)
$$

Sementara:

$$
c \cdot T(\mathbf{a}) = c \cdot (a_1 + a_2, a_1) = (c(a_1 + a_2), c a_1)
$$

Jadi:

$$
T(c \cdot \mathbf{a}) = c \cdot T(\mathbf{a})
$$

**Syarat homogen terpenuhi**

---

#### Kesimpulan :

Karena transformasi:

$$
T(v_1, v_2) = (v_1 + v_2, v_1)
$$

memenuhi kedua sifat aditif dan homogen, maka dapat disimpulkan bahwa:

$$
\boxed{T \text{ adalah transformasi linier.}}
$$

# Definisi Eigenvector dan Eigenvalue

Eigenvector / Eigenvalue yaitu skalar yang digunakan untuk mengubah atau meregangkan vektor eigen.

Lebih tepatnya, eigenvalue (nilai eigen) adalah sebuah skalar yang menyatakan seberapa besar sebuah eigenvector (vektor eigen) berubah arah atau panjangnya setelah suatu transformasi linier dilakukan.

Sementara itu, eigenvector adalah vektor khusus yang arahnya tidak berubah setelah dikenai transformasi linier (misalnya oleh sebuah matriks), meskipun panjangnya bisa berubah.

Secara matematis, hubungan antara matriks $A$, eigenvector $\vec{v}$, dan eigenvalue $\lambda$ dinyatakan dalam persamaan:

$$
A \vec{v} = \lambda \vec{v}
$$

Artinya, ketika matriks $A$ diaplikasikan ke vektor $\vec{v}$, hasilnya adalah vektor yang searah dengan $\vec{v}$ itu sendiri, tetapi diperbesar atau diperkecil oleh faktor $\lambda$.

---
**contoh**

Sebuah panah yang digambarkan di bidang datar, dan panah tersebut mewakili sebuah vektor. Ketika dilakukan transformasi linier—misalnya rotasi, refleksi, atau penskalaan—umumnya arah dan panjang vektor akan berubah. Namun, ada vektor-vektor tertentu yang setelah dikenai transformasi, arah vektornya tetap sama, hanya panjangnya saja yang berubah (bisa bertambah, berkurang, atau bahkan dibalik arahnya).

Vektor yang tidak berubah arah setelah transformasi inilah yang disebut eigenvector, dan skalar yang menyatakan perubahan panjang (diperbesar, diperkecil, atau dibalik) disebut eigenvalue. Jadi, eigenvalue menunjukkan sejauh mana eigenvector diregangkan atau dikompresi oleh transformasi tersebut.

# Mencari Eigenvalue dan Eigenvector


#### Diketahui matriks:

$$
A = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}
$$

---

#### **Langkah 1: Menentukan Eigenvalue**

#####  Tujuan:

Mencari $\lambda$ yang memenuhi  $\det(A - \lambda I) = 0$

#####  Langkah:

Kurangkan $\lambda$ dari elemen diagonal utama matriks $A$:

$$
A - \lambda I =
\begin{bmatrix}
2 - \lambda & 1 \\
1 & 2 - \lambda
\end{bmatrix}
$$

####  Hitung determinan:

$$
\det(A - \lambda I) = (2 - \lambda)^2 - (1)(1)
= (2 - \lambda)^2 - 1
$$

####  Selesaikan persamaan:

$$
(2 - \lambda)^2 = 1
\Rightarrow 2 - \lambda = \pm 1
$$

$$
\lambda = 2 \pm 1 \Rightarrow \lambda = 1 \text{ dan } \lambda = 3
$$

 **hasilnya :**

$$
\lambda_1 = 1,\quad \lambda_2 = 3
$$

---


## **Langkah 2: Menentukan Eigenvector**

#### Untuk $\lambda = 1$

1. Susun matriks $A - \lambda I$:

   $$
   A - I = \begin{bmatrix} 2 - 1 & 1 \\ 1 & 2 - 1 \end{bmatrix} = \begin{bmatrix} 1 & 1 \\ 1 & 1 \end{bmatrix}
   $$

2. Kalikan dengan vektor $\vec{v} = \begin{bmatrix} x \\ y \end{bmatrix}$ dan setarakan dengan nol:

   $$
   \begin{bmatrix} 1 & 1 \\ 1 & 1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} x + y \\ x + y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}
   $$

3. Diperoleh persamaan:

   $$
   x + y = 0 \Rightarrow y = -x
   $$

Maka, **eigenvector untuk $\lambda = 1$** adalah:

$$
\vec{v}_1 = \begin{bmatrix} 1 \\ -1 \end{bmatrix} \quad \text{(atau kelipatannya seperti } \begin{bmatrix} 2 \\ -2 \end{bmatrix})
$$

---

###  Untuk $\lambda = 3$

1. Susun matriks $A - 3I$:

   $$
   A - 3I = \begin{bmatrix} 2 - 3 & 1 \\ 1 & 2 - 3 \end{bmatrix} = \begin{bmatrix} -1 & 1 \\ 1 & -1 \end{bmatrix}
   $$

2. Kalikan dengan vektor $\vec{v} = \begin{bmatrix} x \\ y \end{bmatrix}$:

   $$
   \begin{bmatrix} -1 & 1 \\ 1 & -1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} -x + y \\ x - y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}
   $$

3. Diperoleh persamaan:

   $$
   -x + y = 0 \Rightarrow y = x
   $$

 Maka, **eigenvector untuk $\lambda = 3$** adalah:

$$
\vec{v}_2 = \begin{bmatrix} 1 \\ 1 \end{bmatrix} \quad \text{(atau kelipatannya seperti } \begin{bmatrix} 2 \\ 2 \end{bmatrix})
$$

### Tujuan:

Mengubah panjang eigenvector menjadi **1 satuan** agar menjadi **vektor ortonormal**.

---

$$

---

####  Hitung panjang vektor:

$$
\| \vec{v}_2 \| = \sqrt{1^2 + 1^2} = \sqrt{2}
$$

####  Normalisasi:

$$
\hat{v}_2 = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$

## *2. Basis dan Dot Product*

Basis adalah vektor-vektor acuan yang digunakan untuk membentuk semua vektor lain di dalam suatu ruang.

Tegak lurus atau tidaknya dua vektor tidak bergantung pada panjang (besar) vektor tersebut.
Yang menentukan apakah dua vektor tegak lurus adalah hasil dari dot product (perkalian titik). Jika dot product-nya nol (0), maka sudut antara dua vektor itu adalah 90°, alias tegak lurus.

### *Contoh Basis 2D (dua dimensi)*
Basis standar di ruang 2D

$$

\end{bmatrix}
\quad

- hitung *dot product* (perkalian titik) dari dua vektor ini:

$$
\begin{bmatrix}

0
\end{bmatrix}
\cdot
\begin{bmatrix}
0 \\
1
\end{bmatrix}
= (1×0) + (0×1) = 0
$$

Karena hasilnya 0, ini berarti *kedua vektor saling tegak lurus* (sudutnya 90°).

# *Cross Product (Perkalian Silang)*
## Definisi Hasil Kali Silang
**Hasil kali silang (cross product)** adalah operasi pada dua vektor dalam ruang tiga dimensi yang menghasilkan sebuah vektor baru yang tegak lurus terhadap kedua vektor asal. Vektor hasil ini memiliki arah yang ditentukan oleh kaidah tangan kanan dan besarnya sama dengan luas jajar genjang yang dibentuk oleh kedua vektor tersebut.


Misalkan:

$$
\vec{u} = (u_1, u_2, u_3),  dan \quad \vec{v} = (v_1, v_2, v_3) 
$$  adalah vektor-vektor di R³. **Hasil kali silang** dari u dan v, dinotasikan sebagai u x v, adalah vektor:



$$
\vec{v} \times \vec{u} = 
\begin{bmatrix}
u_2v_3 - u_3v_2 \\
-(u_1v_3 - u_3v_1) \\
u_1v_2 - u_2v_1
\end{bmatrix}
$$

---
##  menghitung hasil kali silang menggunakan determinan 
Pertama, kita membentuk array 3 X 3 seperti yang ditunjukkan di bawah ini.



$$
\vec{u} \times \vec{v} = 
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
u_1 & u_2 & u_3 \\
v_1 & v_2 & v_3 \\
\end{vmatrix}
$$

$$
= \hat{i}(u_2v_3 - u_3v_2) - \hat{j}(u_1v_3 - u_3v_1) + \hat{k}(u_1v_2 - u_2v_1)
$$

$$
= (u_2v_3 - u_3v_2)\hat{i} - (u_1v_3 - u_3v_1)\hat{j} + (u_1v_2 - u_2v_1)\hat{k}
$$
---
## contoh 


Diberikan dua vektor:

$$
\vec{u} = \langle 2,\ -1,\ 4 \rangle, \quad \vec{v} = \langle 3,\ 2,\ 5 \rangle
$$

Hitung perkalian silang:

$$
\vec{u} \times \vec{v} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
2 & -1 & 4 \\
3 & 2 & 5 \\
\end{vmatrix}
$$

Langkah-langkah determinan:

1. Untuk komponen i:

$$
\hat{i} \left| \begin{matrix}
-1 & 4 \\
2 & 5 \\
\end{matrix} \right|
= (-1)(5) - (4)(2)\ \hat{i}
= -5 - 8 = -13\hat{i}
$$

2. Untuk komponen j:

$$
-\hat{j} \left| \begin{matrix}
2 & 4 \\
3 & 5 \\
\end{matrix} \right|
= -((2)(5) - (4)(3))\ \hat{j}
= -(10 - 12) = 2\hat{j}
$$

3. Untuk komponen k:

$$
\hat{k} \left| \begin{matrix}
2 & -1 \\
3 & 2 \\
\end{matrix} \right|
= (2)(2) - (-1)(3)\ \hat{k}
= 4 + 3 = 7\hat{k}
$$

Sehingga:

$$
\vec{u} \times \vec{v} = -13\hat{i} + 2\hat{j} + 7\hat{k}
= \begin{bmatrix} -13 \\ 2 \\ 7 \end{bmatrix}
$$

## Sifat-sifat Perkalian silang

1.  𝑢⃗ × 𝑣⃗ = −(𝑣⃗ × 𝑢⃗) Anticommutative Property

5.  𝑢⃗ × 𝑢⃗ = 0⃗

6.  𝑢⃗ × 0⃗ = 0⃗

7.  𝑢⃗ · (𝑣⃗ × 𝑤⃗) = (𝑢⃗ × 𝑣⃗) · 𝑤⃗  Triple Scalar Product

## Implementasi
Luas jajaran genjang secara geometri adalah A = b × h, dengan b sebagai alas dan h sebagai tinggi. Jika jajaran genjang dibentuk oleh dua vektor u⃗ dan v⃗ dari titik pangkal yang sama, maka tinggi h bisa dinyatakan dengan trigonometri sebagai:

h = ‖u⃗‖ sin(θ)

Sehingga luasnya menjadi:

A = ‖u⃗‖‖v⃗‖ sin(θ) = ‖u⃗ × v⃗‖

Jadi, luas jajaran genjang yang dibentuk dua vektor sama dengan panjang vektor hasil cross product-nya.

Hitung cross product:

$$
\vec{u} \times \vec{v} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
2 & 0 & 0 \\
0 & 3 & 0
\end{vmatrix}
= \hat{i}(0 \cdot 0 - 0 \cdot 3) - \hat{j}(2 \cdot 0 - 0 \cdot 0) + \hat{k}(2 \cdot 3 - 0 \cdot 0)
$$

$$
= \hat{i}(0) - \hat{j}(0) + \hat{k}(6) = 6\hat{k}
$$

$$
\Rightarrow \vec{u} \times \vec{v} = \begin{bmatrix} 0 \\ 0 \\ 6 \end{bmatrix}
$$

Hitung panjangnya:

$$
A = \| \vec{u} \times \vec{v} \| = \sqrt{0^2 + 0^2 + 6^2} = \sqrt{36} = 6
$$

$$
\boxed{Luas = 6}
$$

# Apa itu SVD
**SVD** adalah singkatan dari Singular Value Decomposition, yang dalam bahasa Indonesia berarti Dekomposisi Nilai Singular. Dalam aljabar linier, SVD adalah faktorisasi matriks yang menguraikan matriks menjadi tiga matriks lain yang lebih sederhana.

## Kegunaan SVD
- Reduksi Dimensi
- Kompresi Data
- Pemrosesan Bahasa
- Topik Modeling (Modeling SVD)

## Bagaimana formula SVD
Berikut adalah **rumus dasar dari Singular Value Decomposition (SVD)**:

$$
A = U \Sigma V^T
$$

Dengan penjelasan:

* $A$ adalah matriks asli berukuran $m \times n$
* $U$ adalah matriks ortogonal berukuran $m \times m$, kolom-kolomnya adalah **vektor singular kiri**
* $\Sigma$ adalah matriks diagonal berukuran $m \times n$, elemen diagonalnya adalah **nilai singular** $\sigma_1, \sigma_2, ..., \sigma_r$, dengan $r$ adalah rank dari $A$
* $V^T$ adalah transpose dari matriks ortogonal $V$ berukuran $n \times n$, kolom-kolom $V$ adalah **vektor singular kanan**


### **Rumus dalam Bentuk Penjumlahan Rank-k:**

$$
A \approx A_k = \sum_{i=1}^{k} \sigma_i \, \vec{u}_i \vec{v}_i^T
$$

* $k$: jumlah nilai singular terbesar yang diambil untuk pendekatan
* $\sigma_i$: nilai singular ke-$i$
* $\vec{u}_i$: vektor singular kiri ke-$i$
* $\vec{v}_i$: vektor singular kanan ke-$i$

Rumus ini sering digunakan dalam **reduksi dimensi** dan **topik modeling**, dengan mengambil hanya beberapa nilai singular terbesar (bukan seluruh nilai).

---

### contoh soal
### Lakukan dekomposisi SVD (Singular Value Decomposition) terhadap matriks:
$$
A = \begin{bmatrix} 3 & 1 \\ 1 & 3 \end{bmatrix}
$$

### Langkah-langkah:

SVD menyatakan:

$$
A = U \Sigma V^T
$$

di mana:

* $U$ dan $V$ adalah matriks ortogonal
* $\Sigma$ adalah matriks diagonal berisi **singular value** (akar dari eigenvalue matriks $A^TA$)

---

### 1. Hitung $A^TA$

$$
A^T = \begin{bmatrix} 3 & 1 \\ 1 & 3 \end{bmatrix}^T = \begin{bmatrix} 3 & 1 \\ 1 & 3 \end{bmatrix}
$$

$$
A^TA = \begin{bmatrix} 3 & 1 \\ 1 & 3 \end{bmatrix} \begin{bmatrix} 3 & 1 \\ 1 & 3 \end{bmatrix} = 
\begin{bmatrix}
3×3 + 1×1 & 3×1 + 1×3 \\
1×3 + 3×1 & 1×1 + 3×3
\end{bmatrix} = 
\begin{bmatrix}
10 & 6 \\
6 & 10
\end{bmatrix}
$$

---

### 2. Cari eigenvalue dan eigenvector dari $A^TA$

Misal $M = \begin{bmatrix} 10 & 6 \\ 6 & 10 \end{bmatrix}$, kita cari $\lambda$ dari:

$$
\det(M - \lambda I) = 0
$$

$$
\begin{vmatrix} 10 - \lambda & 6 \\ 6 & 10 - \lambda \end{vmatrix} = (10-\lambda)^2 - 36 = 0
$$

$$
(10 - \lambda)^2 = 36 \Rightarrow 10 - \lambda = \pm6 \Rightarrow \lambda = 4, 16
$$

Jadi eigenvalue-nya: $\lambda_1 = 16, \lambda_2 = 4$

---

### 3. Cari vektor eigen $v_1, v_2$ dari $A^TA$

#### Untuk $\lambda = 16$:

$$
(M - 16I) = \begin{bmatrix} -6 & 6 \\ 6 & -6 \end{bmatrix}
$$

Solusi: $v_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$

#### Untuk $\lambda = 4$:

$$
(M - 4I) = \begin{bmatrix} 6 & 6 \\ 6 & 6 \end{bmatrix}
$$

Solusi: $v_2 = \begin{bmatrix} 1 \\ -1 \end{bmatrix}$

Lalu normalkan:

$$
v_1 = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix},\quad
v_2 = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ -1 \end{bmatrix}
$$

Maka:

$$
V = \begin{bmatrix}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
\end{bmatrix}
$$

---

### 4. Cari singular value (akar dari eigenvalue)

$$
\sigma_1 = \sqrt{16} = 4, \quad \sigma_2 = \sqrt{4} = 2
\Rightarrow \Sigma = \begin{bmatrix} 4 & 0 \\ 0 & 2 \end{bmatrix}
$$

---

### 5. Cari matriks $U$

$$
u_i = \frac{1}{\sigma_i} A v_i
$$

#### Untuk $v_1 = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix}$:

$$
Av_1 = A \cdot \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix} = \frac{1}{\sqrt{2}} \begin{bmatrix} 3+1 \\ 1+3 \end{bmatrix} = \frac{1}{\sqrt{2}} \begin{bmatrix} 4 \\ 4 \end{bmatrix}
\Rightarrow u_1 = \frac{1}{4} \cdot \frac{1}{\sqrt{2}} \begin{bmatrix} 4 \\ 4 \end{bmatrix} = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$

#### Untuk $v_2 = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ -1 \end{bmatrix}$:

$$
Av_2 = \frac{1}{\sqrt{2}} \begin{bmatrix} 3 -1 \\ 1 -3 \end{bmatrix} = \frac{1}{\sqrt{2}} \begin{bmatrix} 2 \\ -2 \end{bmatrix}
\Rightarrow u_2 = \frac{1}{2} \cdot \frac{1}{\sqrt{2}} \begin{bmatrix} 2 \\ -2 \end{bmatrix} = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ -1 \end{bmatrix}
$$

Maka:

$$
U = \begin{bmatrix}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
\end{bmatrix}
$$

---

### Jadi, hasil SVD dari $A = \begin{bmatrix} 3 & 1 \\ 1 & 3 \end{bmatrix}$ adalah:

$$
A = U \Sigma V^T
$$

dengan:

$$
U = \begin{bmatrix}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
\end{bmatrix},\quad
\Sigma = \begin{bmatrix}
4 & 0 \\
0 & 2
\end{bmatrix},\quad
V^T = \begin{bmatrix}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
\end{bmatrix}
$$

## Tugas !
## Lakukan dekomposisi SVD (Singular Value Decomposition) terhadap matriks:

$$
A = \begin{bmatrix} 4 & 1 \\ 2 & 7 \\ 1 & 4 \end{bmatrix}
$$

---

### Langkah-langkah:

SVD menyatakan:

$$
A = U \Sigma V^T
$$

di mana:

* $U$ dan $V$ adalah matriks ortonormal
* $\Sigma$ adalah matriks diagonal yang memuat **singular value** (akar dari eigenvalue matriks $A^T A$)

---

### 1. Hitung $A^TA$

$$
A = \begin{bmatrix}
4 & 1 \\
2 & 7 \\
1 & 4
\end{bmatrix}, \quad
A^T = \begin{bmatrix}
4 & 2 & 1 \\
1 & 7 & 4
\end{bmatrix}
$$


$A^TA$ = \begin{bmatrix}
4 & 2 & 1 \\
1 & 7 & 4
\end{bmatrix}
\begin{bmatrix}
4 & 1 \\
2 & 7 \\
1 & 4
\end{bmatrix}

=
\begin{bmatrix}
4×4 + 2×2 + 1×1 & 4×1 + 2×7 + 1×4 \\
1×4 + 7×2 + 4×1 & 1×1 + 7×7 + 4×4
\end{bmatrix}

=
\begin{bmatrix}
16 + 4 + 1 & 4 + 14 + 4 \\
4 + 14 + 4 & 1 + 49 + 16
\end{bmatrix}

=
\begin{bmatrix}
21 & 22 \\
22 & 66
\end{bmatrix}


---


### 2. Cari eigenvalue dari $A^T A$

Misal $M = \begin{bmatrix} 21 & 22 \\ 22 & 66 \end{bmatrix}$, kita cari nilai $\lambda$ dari:

$$
\det(M - \lambda I) = 0
\Rightarrow
\begin{vmatrix}
21 - \lambda & 22 \\
22 & 66 - \lambda
\end{vmatrix}
= (21 - \lambda)(66 - \lambda) - 22^2 = 0
$$

$$
= \lambda^2 - 87\lambda + 902 = 0
\Rightarrow \lambda_1 = \frac{87 + \sqrt{3961}}{2}, \quad \lambda_2 = \frac{87 - \sqrt{3961}}{2}
$$

---

### 3. Cari vektor eigen $v_1, v_2$

#### Untuk $\lambda_1$:

$$
(M - \lambda_1 I)v = 0
\Rightarrow \text{hasilkan } v_1 \text{ (vektor eigen dari } \lambda_1\text{), lalu normalkan}
$$

#### Untuk $\lambda_2$:

$$
(M - \lambda_2 I)v = 0
\Rightarrow \text{hasilkan } v_2 \text{ (vektor eigen dari } \lambda_2\text{), lalu normalkan}
$$

Setelah itu:

$$
V = \begin{bmatrix} v_1 & v_2 \end{bmatrix}
$$

---

### 4. Hitung singular value

$$
\sigma_1 = \sqrt{\lambda_1}, \quad \sigma_2 = \sqrt{\lambda_2}
\Rightarrow
\Sigma = \begin{bmatrix}
\sigma_1 & 0 \\
0 & \sigma_2 \\
0 & 0
\end{bmatrix}
$$

---

### 5. Hitung matriks $U$

Gunakan:

$$
u_i = \frac{1}{\sigma_i} A v_i
$$

Untuk $i = 1,2$:

$$
u_1 = \frac{1}{\sigma_1} A v_1, \quad
u_2 = \frac{1}{\sigma_2} A v_2
$$

Lalu ambil $u_3$ sebagai vektor ortonormal terhadap $u_1$ dan $u_2$, misalnya dengan Gram-Schmidt.

Maka:

$$
U = \begin{bmatrix}
u_1 & u_2 & u_3
\end{bmatrix}
$$

---

### Jadi, hasil dekomposisi SVD dari $A$ adalah:

$$
A = U \Sigma V^T
$$
